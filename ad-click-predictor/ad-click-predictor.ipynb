# Creating model using Logistic Regression

**Spliting the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Train and fit the model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations

```python
predictions = logmodel.predict(X_test)
```

**Creating a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
